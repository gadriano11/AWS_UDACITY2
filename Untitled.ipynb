# FORD DATA SET

```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
ford_df = pd.read_csv("ford.csv")
print(ford_df.head())
```

```python
# Check the distribution of at least two univariate variables
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Distribution of 'price'
sns.histplot(ford_df['price'], kde=True, ax=axes[0])
axes[0].set_title('Distribution of Car Prices')
axes[0].set_xlabel('Price')

# Distribution of 'mileage'
sns.histplot(ford_df['mileage'], kde=True, ax=axes[1])
axes[1].set_title('Distribution of Car Mileage')
axes[1].set_xlabel('Mileage')

plt.tight_layout()
plt.show()
```

The distribution of car prices shows a right-skewed distribution with most prices concentrated below 20,000.

The distribution of car mileage also shows a right-skewed distribution with most cars having a mileage below 30,000.

```python
# Find the relationship between two numerical variables: 'price' and 'mileage'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='mileage', y='price', data=ford_df)
plt.title('Relationship between Mileage and Price')
plt.xlabel('Mileage')
plt.ylabel('Price')
plt.show()
```

There is an observable negative relationship between mileage and price, indicating that higher mileage cars tend to have lower prices.

```python
# Find the relationship between a numerical variable and a categorical variable: 'price' and 'transmission'
plt.figure(figsize=(10, 6))
sns.boxplot(x='transmission', y='price', data=ford_df)
plt.title('Relationship between Transmission Type and Price')
plt.xlabel('Transmission')
plt.ylabel('Price')
plt.show()
```

The boxplot shows that cars with automatic transmissions tend to have higher prices compared to manual transmissions.

```python
# Find the relationship between two categorical variables: 'transmission' and 'fuelType'
plt.figure(figsize=(10, 6))
sns.countplot(x='transmission', hue='fuelType', data=ford_df)
plt.title('Relationship between Transmission Type and Fuel Type')
plt.xlabel('Transmission')
plt.ylabel('Count')
plt.show()
```

The count plot indicates that petrol cars are more common than diesel cars across both transmission types. However, there are more petrol cars with manual transmission compared to automatic transmission.

```python
#  [Bonus] Find the relationship between more than two variables: 'price', 'engineSize', and 'transmission'
plt.figure(figsize=(12, 8))
sns.scatterplot(x='engineSize', y='price', hue='transmission', data=ford_df, palette='viridis')
plt.title('Relationship between Engine Size, Price, and Transmission Type')
plt.xlabel('Engine Size')
plt.ylabel('Price')
plt.legend(title='Transmission')
plt.show()
```

The scatter plot reveals that, in general, cars with larger engine sizes tend to be priced higher. Additionally, cars with automatic transmissions appear to have higher prices across various engine sizes compared to manual transmissions.
